1. Importación de librerías

```python
import pandas as pd
from sklearn.decomposition import PCA
```

2. Lectura de archivo CSV y generación de Dataframe

```python
EmpleadosAttrition = pd.read_csv('/content/drive/MyDrive/ing_caracteristicas/empleadosRETO.csv')
```

3. Eliminación de características no significativas

```python
EmpleadosAttrition.drop(['EmployeeCount','EmployeeNumber','StandardHours','Over18'], axis=1, inplace=True)
EmpleadosAttrition.head()
```

4 y 5. Calcular años del empleado trabajando para la compañía. En el registro 229 hay una fecha incorrecta (02/30/2012). Se reemplaza por 02/29/2012

```python
EmpleadosAttrition.loc[229, 'HiringDate']='02/29/2012'
EmpleadosAttrition['Year'] = pd.DatetimeIndex(EmpleadosAttrition['HiringDate']).year.astype(int)
EmpleadosAttrition.head()

```

6. Calcular cantidad de años trabajando para la Compañía por empleado

```python
EmpleadosAttrition['YearsatCompany'] = 2018 - EmpleadosAttrition['Year']
EmpleadosAttrition.head()
```

8. Renombrar la variable **DistanceFromHome** a **DistanceFromHome_km**

```python
EmpleadosAttrition.rename(columns={'DistanceFromHome': 'DistanceFromHome_km'},inplace=True)
EmpleadosAttrition.head()
```

7 y 9. Crear nueva variable **DistanceFromHome** entera (eliminación de unidades (km) y tranformación de valores de typo cadena (**string**) a tipo entero (**int**)) la cual queda al final del Dataframe

```python

EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome_km'].replace(' km','', regex=True)
EmpleadosAttrition['DistanceFromHome'] = EmpleadosAttrition['DistanceFromHome'].astype(int)
EmpleadosAttrition.head()
```

10. Eliminación de las columnas **Year**, **HiringDate** y **DistanceFromHome_km**


```python
EmpleadosAttrition.drop(['Year','HiringDate','DistanceFromHome_km'], axis=1, inplace=True)
EmpleadosAttrition.head()
```

11. Crear Dataframe SueldoPromedioDepto el cual incluya la variable MonthlyIncome

```python
SueldoPromedio = EmpleadosAttrition.groupby('Department')['MonthlyIncome'].mean()
SueldoPromedioDepto = pd.DataFrame(SueldoPromedio)
SueldoPromedioDepto = SueldoPromedioDepto.rename(columns={'MonthlyIncome': 'SueldoPromedio'})
SueldoPromedioDepto
```

12.a Escalar la vaiable MonthlyIncome para que tenga valores entre 0 y 1 (Escalamiento Min-Max)

```python
EmpleadosAttrition['MonthlyIncome_Esc'] = (EmpleadosAttrition['MonthlyIncome']-EmpleadosAttrition['MonthlyIncome'].min()) / (max(EmpleadosAttrition['MonthlyIncome'])-EmpleadosAttrition['MonthlyIncome'].min())
EmpleadosAttrition.drop(['MonthlyIncome'], axis=1, inplace=True)
EmpleadosAttrition.head()
```

12.b Se escala el resto de las variables numéricas

```python
EmpleadosAttrition['YearsSinceLastPromotion_Esc'] = (EmpleadosAttrition['YearsSinceLastPromotion']-EmpleadosAttrition['YearsSinceLastPromotion'].min()) / (max(EmpleadosAttrition['YearsSinceLastPromotion'])-EmpleadosAttrition['YearsSinceLastPromotion'].min())
EmpleadosAttrition['WorkLifeBalance_Esc'] = (EmpleadosAttrition['WorkLifeBalance']-EmpleadosAttrition['WorkLifeBalance'].min()) / (max(EmpleadosAttrition['WorkLifeBalance'])-EmpleadosAttrition['WorkLifeBalance'].min())
EmpleadosAttrition['TrainingTimesLastYear_Esc'] = (EmpleadosAttrition['TrainingTimesLastYear']-EmpleadosAttrition['TrainingTimesLastYear'].min()) / (max(EmpleadosAttrition['TrainingTimesLastYear'])-EmpleadosAttrition['TrainingTimesLastYear'].min())
EmpleadosAttrition['RelationshipSatisfaction_Esc'] = (EmpleadosAttrition['RelationshipSatisfaction']-EmpleadosAttrition['RelationshipSatisfaction'].min()) / (max(EmpleadosAttrition['RelationshipSatisfaction'])-EmpleadosAttrition['RelationshipSatisfaction'].min())
EmpleadosAttrition['PerformanceRating_Esc'] = (EmpleadosAttrition['PerformanceRating']-EmpleadosAttrition['PerformanceRating'].min()) / (max(EmpleadosAttrition['PerformanceRating'])-EmpleadosAttrition['PerformanceRating'].min())
EmpleadosAttrition['NumCompaniesWorked_Esc'] = (EmpleadosAttrition['NumCompaniesWorked']-EmpleadosAttrition['NumCompaniesWorked'].min()) / (max(EmpleadosAttrition['NumCompaniesWorked'])-EmpleadosAttrition['NumCompaniesWorked'].min())
EmpleadosAttrition['Education_Esc'] = (EmpleadosAttrition['Education']-EmpleadosAttrition['Education'].min()) / (max(EmpleadosAttrition['Education'])-EmpleadosAttrition['Education'].min())
EmpleadosAttrition['PercentSalaryHike_Esc'] = (EmpleadosAttrition['PercentSalaryHike']-EmpleadosAttrition['PercentSalaryHike'].min()) / (max(EmpleadosAttrition['PercentSalaryHike'])-EmpleadosAttrition['PercentSalaryHike'].min())
EmpleadosAttrition['Age_Esc'] = (EmpleadosAttrition['Age']-EmpleadosAttrition['Age'].min()) / (max(EmpleadosAttrition['Age'])-EmpleadosAttrition['Age'].min())
EmpleadosAttrition['DistanceFromHome_Esc'] = (EmpleadosAttrition['DistanceFromHome']-EmpleadosAttrition['DistanceFromHome'].min()) / (max(EmpleadosAttrition['DistanceFromHome'])-EmpleadosAttrition['DistanceFromHome'].min())
EmpleadosAttrition['EnvironmentSatisfaction_Esc'] = (EmpleadosAttrition['EnvironmentSatisfaction']-EmpleadosAttrition['EnvironmentSatisfaction'].min()) / (max(EmpleadosAttrition['EnvironmentSatisfaction'])-EmpleadosAttrition['EnvironmentSatisfaction'].min())
EmpleadosAttrition['JobInvolvement_Esc'] = (EmpleadosAttrition['JobInvolvement']-EmpleadosAttrition['JobInvolvement'].min()) / (max(EmpleadosAttrition['JobInvolvement'])-EmpleadosAttrition['JobInvolvement'].min())
EmpleadosAttrition['JobLevel_Esc'] = (EmpleadosAttrition['JobLevel']-EmpleadosAttrition['JobLevel'].min()) / (max(EmpleadosAttrition['JobLevel'])-EmpleadosAttrition['JobLevel'].min())
EmpleadosAttrition['JobSatisfaction_Esc'] = (EmpleadosAttrition['JobSatisfaction']-EmpleadosAttrition['JobSatisfaction'].min()) / (max(EmpleadosAttrition['JobSatisfaction'])-EmpleadosAttrition['JobSatisfaction'].min())
EmpleadosAttrition['TotalWorkingYears_Esc'] = (EmpleadosAttrition['TotalWorkingYears']-EmpleadosAttrition['TotalWorkingYears'].min()) / (max(EmpleadosAttrition['TotalWorkingYears'])-EmpleadosAttrition['TotalWorkingYears'].min())
EmpleadosAttrition['YearsInCurrentRole_Esc'] = (EmpleadosAttrition['YearsInCurrentRole']-EmpleadosAttrition['YearsInCurrentRole'].min()) / (max(EmpleadosAttrition['YearsInCurrentRole'])-EmpleadosAttrition['YearsInCurrentRole'].min())
EmpleadosAttrition['YearsatCompany_Esc'] = (EmpleadosAttrition['YearsatCompany']-EmpleadosAttrition['YearsatCompany'].min()) / (max(EmpleadosAttrition['YearsatCompany'])-EmpleadosAttrition['YearsInCurrentRole'].min())
EmpleadosAttrition.drop(['WorkLifeBalance','YearsSinceLastPromotion','TrainingTimesLastYear','RelationshipSatisfaction','PerformanceRating','NumCompaniesWorked','Education','PercentSalaryHike','DistanceFromHome','Age','EnvironmentSatisfaction','JobInvolvement','JobLevel','JobSatisfaction','TotalWorkingYears','YearsInCurrentRole','YearsatCompany'], axis=1, inplace=True)
EmpleadosAttrition.head()
```

13. Convertir las variables categóricas **BusinessTravel**, **Department**, **EducationField**, **Gender**, **JobRole**, **MaritalStatus** y **Attrition** a numéricas. Se agrega variable **Overtime**. Se concatenan los Dataframes creados al Dataframe **EmpleadosAttition** y se eliminan las variables categóricas originales (no numéricas) así como aquellas variables binarias (**Gender**, **Overtime** y **Attrition**)

```python
EmpleadosAttrition_BT = pd.get_dummies(EmpleadosAttrition.BusinessTravel, prefix = 'BusinessTravel',dtype=int)
EmpleadosAttrition_Dept = pd.get_dummies(EmpleadosAttrition.Department, prefix = 'Department',dtype=int)
EmpleadosAttrition_EF = pd.get_dummies(EmpleadosAttrition.EducationField, prefix = 'EducationField',dtype=int)
EmpleadosAttrition_Gdr = pd.get_dummies(EmpleadosAttrition.Gender, prefix = 'Gender',dtype=int)
EmpleadosAttrition_JR = pd.get_dummies(EmpleadosAttrition.JobRole, prefix = 'JobRole',dtype=int)
EmpleadosAttrition_MS = pd.get_dummies(EmpleadosAttrition.MaritalStatus, prefix = 'MaritalStatus',dtype=int)
EmpleadosAttrition_OT = pd.get_dummies(EmpleadosAttrition.OverTime, prefix = 'OverTime',dtype=int)
EmpleadosAttrition_Att = pd.get_dummies(EmpleadosAttrition.Attrition, prefix = 'Attrition',dtype=int)
EmpleadosAttrition = pd.concat([EmpleadosAttrition,EmpleadosAttrition_BT,EmpleadosAttrition_Dept,EmpleadosAttrition_EF,EmpleadosAttrition_Gdr,EmpleadosAttrition_JR,EmpleadosAttrition_MS,EmpleadosAttrition_OT, EmpleadosAttrition_Att], axis=1)
EmpleadosAttrition = EmpleadosAttrition.drop(['BusinessTravel','Department','EducationField','Gender','JobRole','MaritalStatus','OverTime','Attrition'], axis=1)
EmpleadosAttrition = EmpleadosAttrition.drop(['Gender_Male','OverTime_Yes','Attrition_Yes'], axis=1)
EmpleadosAttrition.head()
```

14. Calcular la correlación entre las variables del Dataframe.

```python
print(EmpleadosAttrition.corr())
```

15. Eliminar variables con correlación menor a 0.1 respecto a variable de salida **Attrition**. Crear el Dataframe **EmpleadosAttritionFinal** con las variiables significativas.

```python
EmpleadosAttritionFinal = EmpleadosAttrition.drop(['Education_Esc','NumCompaniesWorked_Esc','PercentSalaryHike_Esc','PerformanceRating_Esc','JobRole_Manufacturing Director','JobRole_Research Scientist','JobRole_Sales Executive','JobRole_Human Resources','JobRole_Manager','MaritalStatus_Married','RelationshipSatisfaction_Esc','TrainingTimesLastYear_Esc','WorkLifeBalance_Esc','YearsSinceLastPromotion_Esc','Gender_Female','DistanceFromHome_Esc','BusinessTravel_Travel_Frequently','BusinessTravel_Travel_Rarely','Department_Human Resources','Department_Research & Development','Department_Sales','EducationField_Life Sciences','EducationField_Other','EducationField_Human Resources','EducationField_Marketing','EducationField_Medical'], axis=1)
EmpleadosAttritionFinal.head()
```

16. Crear la variable **EmpleadosAttritionPC**A a partir del Dataframe **EmpleadosAttritionFinal**.

```python
pca = PCA()
pca.fit(EmpleadosAttritionFinal)
print("Componentes PCA: ")
print(pca.components_)
print("Varianza Explicada: ")
print(pca.explained_variance_)
print("Razón de la Varianza Explicada: ")
print(pca.explained_variance_ratio_)
EmpleadosAttritionPCA = pca.transform(EmpleadosAttritionFinal)
print("Componentes Principales: ")
EmpleadosAttritionPCA = pd.DataFrame(EmpleadosAttritionPCA)
EmpleadosAttritionPCA.head()
```

17. Agregar el mínimo número de Componentes Principales en columnas del Dataframe **EmpleadosAttritionPCA** que logren explicar el 80% de la varianza, al Dataframe **EmpleadosAttritionFinal**.

```python
EmpleadosAttritionPCA = EmpleadosAttritionPCA.drop(EmpleadosAttritionPCA.columns[[10, 11, 12, 13, 14, 15, 16,17, 18]], axis='columns')
EmpleadosAttritionFinal = pd.concat([EmpleadosAttritionFinal,EmpleadosAttritionPCA], axis=1)
EmpleadosAttritionFinal.head()
```

18. Guardar el set de datos **EmpleadosAttritionFinal** en un archivo CSV llamado EmpleadosAttritionFinal.csv.

```python
EmpleadosAttritionFinal.to_csv('EmpleadosAttritionFinal.csv')
```
